```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv("wine.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Normalizing the data
norm_data=scale(data)
```

```python
norm_data
```

Implementing PCA

```python
pca=PCA(n_components=14)
data_pca=pca.fit_transform(norm_data)
data_pca
```

Covariance matrix

```python
pca.components_
```

```python
# amount of variance
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
plt.plot(var1)
```

```python
finaldf=pd.concat([data["Type"],pd.DataFrame(data_pca[:,0:3],columns=["PC1","PC2","PC3"])],axis=1)
finaldf
```

```python
#PCA Visualization
fig=plt.figure(figsize=(10,10))
sns.scatterplot(data=finaldf)
```

Using Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,10))
dendrogram=sch.dendrogram(sch.linkage(norm_data,"complete"))
```

```python
hcClusters=AgglomerativeClustering(n_clusters=3,affinity="euclidean",linkage="ward")
hcClusters
```

```python
y=pd.DataFrame(hcClusters.fit_predict(norm_data),columns=["Cluster ID"])
y["Cluster ID"].value_counts()
```

```python
data1=data.copy()
data1["Cluster ID"]=hcClusters.labels_
data1
```

K Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,6):
  kmeans=KMeans(n_clusters=i,random_state=5)
  kmeans.fit(norm_data)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,6),wcss)
plt.title("elbow curve")
plt.xlabel("---No of Clusters-->")
plt.ylabel("wcss")
plt.show();
```

from the elbow curev we can see no of cluster =3 is good choice

```python
cluster=KMeans(3,random_state=5).fit(norm_data)
cluster
```

```python
data2=data.copy()
data2["Cluster ID"]=cluster.labels_
data2
```
